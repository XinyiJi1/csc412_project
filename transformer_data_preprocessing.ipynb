```python
import numpy as np
import pandas as pd
```

```python
import os, sys

# Open a file
path = "./image_to_vector/image_to_vector2"
dirs = os.listdir( path )

df = pd.DataFrame(columns=range(1280))
# This would print all the files and directories
i = 0
for file in dirs:
    if i >= 4000:
        break
    i = i + 1
    cur = "./image_to_vector/image_to_vector2/" + file
    txt_list = np.loadtxt(cur, dtype=str)
    cur_df = pd.DataFrame(txt_list.reshape(-1, len(txt_list)))
    df = pd.concat([df, cur_df])
    if i % 200 == 0:
        print(i)
df
```

```python
df = df.astype(int)
df
```

```python
df.to_csv('transformer_data.csv', index=False)
```

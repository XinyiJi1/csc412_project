```python
df_label = pd.read_csv('transformer_data_new_label.csv')  
df_label
```

```python
df_train['label'] = list(df_label[df_label.index < 2700]['label'])
df_train
```

```python
df_test = pd.read_csv('test_data_mlp.csv')  
df_test
```

```python
df_test['label'] = list(df_label[(df_label.index >= 2700) & (df_label.index < 3370)]['label'])
df_test
```

```python
df = pd.concat([df_train, df_test])
df = df.reset_index(drop=True)
df
```

```python
df_origin = pd.read_csv('transformer_data_new_label_data.csv')  
df_origin = df_origin[df_origin.index < 3370]
df_origin
```

```python
count_dict = df['label'].value_counts()
category_saved = []
for key in count_dict.keys():
    if count_dict[key] >= 100:
        category_saved.append(key)
category_saved
```

```python
len(category_saved)
```

```python
df = df[df['label'].isin(category_saved)]
df
```

```python
df['label'].to_csv('transformer_data_new_label.csv', index=False)
```

```python
df.to_csv('transformer_data_new_label_data.csv', index=False)
```

```python
image_saved = df.index
image_saved
```

```python
#from bpe import *
#from dvae_sample import *
import os
res = []
ans = 0
image_to_caption_file = open('./image_to_caption.txt', 'r')
image_to_caption_lines = image_to_caption_file.readlines()
for line in image_to_caption_lines:
    
    img_num = int(line[2:7])
    
    if img_num - 1 in image_saved:
        
        ans += 1
        image_captions = line[11:-2]
        image_to_caption_string = image_captions.replace('\\n', '')
        res = res +  [image_to_caption_string]
```

```python
ans
```

```python
res_len = []
for r in res:
    res_len.append(len(r))
```

```python
import matplotlib.pyplot as plt
import numpy as np


plt.hist(res_len, density=True, bins=40, color = 'skyblue')  # density=False would make counts
plt.title("Distribution of caption length")
plt.ylabel('Probability')
plt.xlabel('Length of caption');
```
